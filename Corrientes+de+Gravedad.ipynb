```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from pylab import *

#Función para dos ejes verticales
def two_scales(ax1,time,data1,data2,c1,c2):
    ax2=ax1.twinx()
    ax1.plot(time,data1,color=c1)
    ax1.set_xlabel('Tiempo[s]')
    ax1.set_ylabel('Temperatura [°C]')
    
    ax2.plot(time,data2,color=c2)
    ax2.set_ylabel('Densidad (kg/m³)')
    return ax1, ax2


# Serie de tiempo
t=np.arange(0,112,1)

#Temperatura
T=np.loadtxt('temp_sensor1_exp2_111.txt')


#Salinidad (densidad)
s=np.loadtxt('fgl_2_calibratelimpio.txt')
a=s[:,2]


#Generando la gráfica
figure, ax=plt.subplots()
ax1,ax2=two_scales(ax,t,T,a,'r','b')
ax.set_title('Serie de tiempo')
ax.grid(True)
figure.set_dpi(160) #Tamaño de la imagen



def color_y_axis(ax, color):
    for t in ax.get_yticklabels():
        t.set_color(color)
    return None
color_y_axis(ax1, 'r')
color_y_axis(ax2, 'b')

plt.legend(bbox_to_anchor=(1.1, 1), loc=3, borderaxespad=0.)

plt.show()
```
