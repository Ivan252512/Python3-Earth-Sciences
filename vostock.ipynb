```python
import matplotlib.pyplot as plt
import numpy as np

temperatura=np.loadtxt('vostok.txt',usecols=3)
CO2=np.loadtxt('vostok.txt',usecols=6)

correlacion=np.corrcoef(temperatura,CO2)
coef=np.polyfit(temperatura,CO2,1)
ecuacion=np.poly1d(coef)
y_estimadas=ecuacion(temperatura)
r_cuadrada=correlacion[0,1]*correlacion[0,1]
r2=np.around(r_cuadrada,4)

plt.plot(temperatura,y_estimadas,'-b',linewidth=1,label='Recta de regresión')
#cadena_ec=str(ecuacion)
cadena_r=str(r2)
#plt.text(-56,200,'y='+cadena_ec[2:len(cadena_ec)])
plt.text(-56,210,'R='+cadena_r)

plt.plot(temperatura,CO2,'.c',linestyle='none')
plt.title('Temperatura vs CO2')
plt.xlabel('Temperatura [°C]')
plt.ylabel('CO2 [ppmv]')
plt.legend(loc=0)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

temperatura=np.loadtxt('pleasestop.txt',usecols=0)
espesor=np.loadtxt('pleasestop.txt',usecols=1)

correlacion=np.corrcoef(espesor,temperatura)
coef=np.polyfit(espesor,temperatura,1)
ecuacion=np.poly1d(coef)
y_estimadas=ecuacion(espesor)
r_cuadrada=correlacion[0,1]*correlacion[0,1]
r2=np.around(r_cuadrada,4)

plt.plot(espesor,y_estimadas,'-b',linewidth=1,label='Recta de regresión')
cadena_r=str(r2)
plt.text(7,24.3,'R='+cadena_r)

plt.plot(espesor,temperatura,'.c',linestyle='none')
plt.title('Espesor vs Temperatura')
plt.xlabel('Espesor [mm]')
plt.ylabel('Temperatura [°C]')
plt.legend(loc=0)
plt.grid(True)
plt.show()
```
