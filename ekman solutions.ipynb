```python
import numpy as np
import matplotlib.pyplot as plt
uo=10
d=10
z=np.arange(0,52,0.5)

u=uo*(1-np.exp(-z/d)*np.cos(z/d))
v=uo*(np.exp(-z/d)*np.sin(z/d))

plt.plot(u,z)
plt.plot(v,z)
plt.grid(True)
plt.show()

     
     
     
     
```
