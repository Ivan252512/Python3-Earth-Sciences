```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Ondas de gravedad sin rotación

k=np.arange(0.1,1,0.1) #Número de onda
w=math.sqrt(9.8*20)*k #Relación de dispersión

#Ondas de gravedad con rotación
f=(10^-4)*(10^-4)  #Valor típico del parámetro de Coriolis en latitudes medias
#Considerando l=o
wr=math.sqrt(f+9.8*20)*k

plt.plot(k,w,label='sin rotación')
plt.plot(k,wr, label='con rotación')
plt.title('Ondas de gravedad superficiales')
plt.xlabel('k [1/m]')
plt.ylabel('w [m/s]')
plt.grid(True)
plt.legend(bbox_to_anchor=(1, 1), loc=3, borderaxespad=1)
plt.show()
```
